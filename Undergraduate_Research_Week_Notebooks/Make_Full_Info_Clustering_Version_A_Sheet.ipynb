```python
import pandas as pd
```

```python
full_info_boder_df = pd.read_excel("../Datasets/Version_A_Results.xlsx")
full_info_boder_df.info()
```

```python
boder_clusters_df = pd.read_excel("../Datasets/Boder_Version_A_Disambiguated_Clusters.xlsx")
boder_clusters_df.info()
```

```python
lists = 0
strings = 0

for text in boder_clusters_df["Text"]:
    if isinstance(text, str):
        strings += 1
    elif isinstance(text, list):
        lists += 1

print(f"Number of strings: {strings}")
print(f"Number of lists: {lists}")
```

```python
import ast

boder_clusters_df["Text"] = boder_clusters_df["Text"].apply(ast.literal_eval)
```

```python
lists = 0
strings = 0

for text in boder_clusters_df["Text"]:
    if isinstance(text, str):
        strings += 1
    elif isinstance(text, list):
        lists += 1

print(f"Number of strings: {strings}")
print(f"Number of lists: {lists}")
```

```python
boder_clusters_df["Text"][0]
```

```python
questions = []
clusters = []

for i in range(95):
    for text in boder_clusters_df["Text"][i]:
        questions.append(text)
        clusters.append(boder_clusters_df["Cluster"][i])

print(f"Number of questions: {len(questions)}")
```

```python
file_nums = []
sent_nums = []
manual_indices = []

for question in questions:
    if question in full_info_boder_df["Final Reviewed Text"].tolist():
        slice = full_info_boder_df[full_info_boder_df["Final Reviewed Text"] == question]
        file_nums.append(slice["file_num"].tolist()[0])
        sent_nums.append(slice["sent_num"].tolist()[0])
        manual_indices.append(slice["manual_index"].tolist()[0])
    else:
        file_nums.append(None)
        sent_nums.append(None)
        manual_indices.append(None)
```

```python
# Create a DataFrame with the results
boder_questions_df = pd.DataFrame({
    "Question": questions,
    "Cluster": clusters,
    "File Num": file_nums,
    "Sentence Num": sent_nums,
    "Manual Index": manual_indices
})

boder_questions_df
```

```python
boder_questions_df.to_excel("../Datasets/Boder_Version_A_Clusters_Full_Info.xlsx", index=False)
```
